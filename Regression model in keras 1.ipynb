```python
###Assigning target i.e, strength and it's features
```

```python
features = concrete_data[concrete_data.columns.drop("Strength")]
target = concrete_data["Strength"]
```

```python
###finding out number of columns
```

```python
n_cols = features.shape[1]
```

```python
###Importing the train_test_split and assigning the value to the variables
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size = 0.3)
```

```python
###Importing keras
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
###Defining the regression model function
```

```python
# define regression model
def regression_model():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error', metrics=["accuracy"])
    return model
```

```python
model = regression_model()
```

```python
###Performing the fitting of data
```

```python
history = model.fit(X_test, y_test, epochs = 50, verbose = 1)
```

```python
###To see how much error is present in our prediction
```

```python
import matplotlib.pyplot as plt
```

```python
loss = history.history["loss"]
epochs = range(0,50)
plt.plot(epochs, loss , label = "prediction_loss")
plt.xlabel = "epochs"
plt.ylabel = "loss"
plt.legend()
plt.show()
```
